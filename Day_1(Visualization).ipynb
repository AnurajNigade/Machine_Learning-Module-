# Matplotlib Library

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(1,10)
y = np.arange(11,20)
```

```python
# Ex : 2-D scatter plot 
plt.scatter(x,y,c='g',s=100)    
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Testing data points')
```

```python
# Ex : 2-D  plot 
y = x*x
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Testing data points')
```

```python
# Ex : Subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*--')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')

plt.subplot(2,2,4)
plt.plot(x,y,'go')
         
```

```python
# Ex : 
x = np.arange(1,11)
y = 3*x + 5
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Testing data points')
```
