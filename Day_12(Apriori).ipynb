```python
transaction = []

for i in range(0, 7501):
    transaction.append([str(df.values[i, j]) for j in range(0, 20)])
```

```python
transaction[:1]
```

```python
from apyori import apriori
rules = apriori(transactions=transaction, min_support=0.003, min_confindence=0.2,min_lift=3,min_length=2,max_length=2 )
```

```python
results = list(rules)
```

```python
print(len(results))
```

```python
def inspect(result):
    lhs = [tuple(result[2][0][0])[0] for result in result]
    rhs = [tuple(result[2][0][1])[0] for result in result]
    supports = [result[1] for result in result]
    confidence = [result[2][0][2] for result in result]
    lifts = [result[2][0][3] for result in result]
    return list(zip(lhs,rhs,supports,confidence,lifts))
resultindataframe = pd.DataFrame(inspect(results), columns = ['LHS', 'RHS', 'Support', 'Confidence', 'Lift'])
```
